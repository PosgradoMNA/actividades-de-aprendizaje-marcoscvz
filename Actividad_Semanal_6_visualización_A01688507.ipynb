###*3. Limpia los datos eliminando los registros nulos o rellena con la media de la columna*

```python
#Primeramente se eliminará las filas en donde la Y es nan debido a que esta es la variable de respuesta y es indispensable contar con su valor 
#La tabla de frecuencias que sólo hay 3 missing values

df[['Y']].value_counts(dropna=False)
```

```python
#Se eliminaran las filas en donde la Y en NaN

ndf = df.copy() #Se copia el dataframe
ndf.drop(ndf[ndf.Y.isnull()].index, inplace = True) #Elimina las columnas en donde hay nulos en la variable Y
ndf[['Y']].value_counts(dropna=False)
```

```python
#Se separan las variables categóricas y numéricas
ndf_cat = ndf[['gender', 'education', 'marital_status', 'hry_sep', 'hry_aug', 'hry_jul', 'hry_jun', 'hry_may', 'hry_apl']]
ndf_num = ndf[['amount_credit',	'age',	'amt_sep',	'amt_aug',	'amt_jul',	'amt_jun',	'amt_may', 'amt_apl',	'paid_sep',	'paid_aug',	'paid_jul',	'paid_jun',	'paid_may',	'paid_apl']]
```

```python
#Verificación
print(list((ndf_cat.columns)))
print(list((ndf_num.columns)))
```

```python
#se definen los pipelines para imputar por la moda y la mediana en lugar de usar la media

from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

pipeline_cat = Pipeline(steps = [
    ("imputer", SimpleImputer(strategy='mean'))
])

pipeline_num = Pipeline(steps = [
    ("imputer", SimpleImputer(strategy='median'))
])
```

```python
#Se define la imputación

from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(transformers=[
    ("cat", pipeline_cat, list((ndf_cat.columns))),
    ("num", pipeline_num, list((ndf_num.columns)))
], remainder='passthrough');
```

```python
#Se aplica la imputación a la base

ndf_imp = pd.DataFrame(transformer.fit_transform(ndf),columns=list((ndf.columns)))
ndf_imp.isnull().sum() #Se verifica que ya no haya NaN
```
